```python
def normalize_condition_names(obs):
    import pandas as pd
    ser = pd.Series(
        ["+".join(sorted(x.split("+"))) for x in obs['condition']],
        dtype="category",
        index=obs.index
    )
    obs['condition'] = ser
    return obs
```
