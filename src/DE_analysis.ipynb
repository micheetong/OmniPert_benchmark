```python
# load packages
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt
import scanpy as sc
```

```python
# load adata 
adata = sc.read_h5ad("/Users/mtong/Desktop/benchmark/southard_RPE1_CRISPRi/southard_data/perturb_processed.h5ad")
```

```python
# run differential expression: compare every perturbation condition vs "ctrl"
sc.tl.rank_genes_groups(
    adata,
    groupby="condition",   # column in adata.obs
    reference="ctrl",      # baseline
    method="wilcoxon"       
)

# Plot the results
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```

```python
id_to_name = adata.var['gene_name'].to_dict()
de_df = sc.get.rank_genes_groups_df(adata, group=None)
de_df['gene_name'] = de_df['names'].map(id_to_name)
```
